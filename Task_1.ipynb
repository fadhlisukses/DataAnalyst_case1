```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/fadhlisukses/DataAnalyst_case1/master/Inflasi.csv'
inflasi = pd.read_csv(url)
inflasi
```

```
inflasi.dtypes
```

```
inflasi['Inflasi'] = inflasi['Inflasi'].str.rstrip('%').astype('float') / 100.0
inflasi
```

```
inflasi.dtypes
```

```
inflasi.plot(x='Month', y='Inflasi', kind='line', figsize=(15,10), title="Diagram Garis Data Inflasi di Indonesia dari Tahun Desember 2002 - Juni 2019")
```

Dari diagram garis diatas dapat diketahui bahwa besar inflasi yang ada di Indonesia dari Desember 2002 hingga Juni 2019 sangat flutuatif. Dari grafik tersebut juga dapat diketahui bahwa nilai inflasi tertinggi terjadi sekitar akhir tahun 2005 hingga pertengahan tahun 2006.

```
inflasi.describe()
```

Dari tabel analisa statistika deskripsi di atas dapat diketahui bahwa:
1. Total data = 199
2. Rata-rata besar inflasi = 0.063 (6.3%)
3. Standar deviasi (sebaran data) = 0.033 (3.3%)
   Artinya, lebih dari 95% sebaran data inflasi berjarak sekitar 3 sampai 4 standar deviasi dari nilai rata-rata.
4. Nilai Minimum inflasi (min) = 0%
5. Lower Precentile = 0.038 (3.8%)
6. Median (50% Precentile) = 0.06 (6%)
7. Upper Precentile = 0.072 (7.2%)
8. Nilai Maximum inflasi (max) = 0.183 (18.3%)

```
inflasi['Inflasi'].plot.box()
```

> Boxplot diatas menggambarkan hasil yang sedikit berbeda dari deskriptif analisis sebelumnya karena dalam box plot nilai maksimum dan minimumnya adalah nilai diluar *outliar*.

Dari Boxplot diatas dapat diketahui bahwa:
1. Terdapat beberapa *outlier* di bagian persentase inflasi yang bernilai besar.
2. Adanya outlier di bagian atas boxplot yang disertai dengan whisker bagian atas yang lebih panjang, menunjukkan bahwa distribusi data cenderung menjulur ke arah kanan *(positive skewness)*. Artinya, nilai rata-rata *(mean)* dan nilai tengah *(median)* berada di sebelah kanan dari modus.

```
infLevel = []
for i in inflasi['Inflasi']:
  if i > 0.5:
    infLevel.append('Hiperinflasi')
  elif i >= 0.1:
    infLevel.append('Inflasi Berderap')
  elif i >= 0.03:
    infLevel.append('Inflasi Berjalan')
  else:
    infLevel.append('Inflasi Ringan')
    
inflasi['Inflasi_Level'] = infLevel    
inflasi
```

```
inflasi_level_counts = inflasi['Inflasi_Level'].value_counts()
inflasi_level_counts = pd.DataFrame(inflasi_level_counts)
inflasi_level_counts.rename(columns={'Inflasi_Level':'value_counts'}, inplace=True)
inflasi_level_counts.index.name=('Inflasi_Level')
inflasi_level_counts
```

Dari value_counts pada kolom level inflasi dapat diketahui bahwa secara umum inflasi yang terjadi di Indonesia dari Desember 2002 hingga Juni 2019 bertipe **Inflasi Berjalan**. Artinya, persentase inflasi paling banyak sekitar 3-10%.
